#3- Compare and contrast mutable and immutable objects in Python with examples.
 - Mutable objects are data types whose values can be changed after they are created. This means you can update, delete, or add items without creating a new object. When you modify a mutable object, its memory address (ID) stays the same because the original object is being changed in place.

Examples of mutable objects: list, dict, set, bytearray


 - Immutable objects are data types whose values cannot be changed after they are created. Any operation that seems to "change" the object actually creates a new object in memory with a different address.

Examples of immutable objects: int, float, str, tuple, bool, frozenset




# 4- Discuss the different types of operators in Python and provide examples of how they are used.
 -  In Python, operators are special symbols or keywords used to perform operations on values or variables. There are several types of operators, each with a specific purpose.

 Types of operator in python are :

  1. Arithmetic Operators :
Used for basic mathematical operations.

Example -

          a = 10
          b = 3

    print(a + b)   # Addition → 13
    print(a - b)   # Subtraction → 7
    print(a * b)   # Multiplication → 30
    print(a / b)   # Division → 3.333...
    print(a % b)   # Modulus (remainder) → 1
    print(a // b)  # Floor division → 3
    print(a ** b)  # Exponentiation → 1000


2. Comparison (Relational) Operators
Used to compare two values; returns True or False.

Example -

    print(x == y)   # False
    print(x != y)   # True
    print(x > y)    # False
    print(x < y)    # True
    print(x >= y)   # False
    print(x <= y)   # Tr


3. Assignment Operators
Used to assign values to variables, with variations that perform operations and then assign.

Examples:    

     a = 5
    a += 2   # a = a + 2 → 7
    a *= 3   # a = a * 3 → 21
    a -= 4   # a = a - 4 → 17
    a /= 2   # a = a / 2 → 8.5

4. Logical Operators
Used to combine conditional statements.

Examples:

    x = True
    y = False

    print(x and y)   # False
    print(x or y)    # True
    print(not x)     # False

5. Bitwise Operators
Work on bits (binary representations).

Examples:

    a = 5     # 0101
    b = 3     # 0011

    print(a & b)   # AND → 1 (0001)
    print(a | b)   # OR → 7 (0111)
    print(a ^ b)   # XOR → 6 (0110)
    print(~a)      # NOT → -6
    print(a << 1)  # Left shift → 10 (1010)
    print(a >> 1)  # Right shift → 2 (0010)

6. Membership Operators
Check for membership in sequences like lists, strings, etc.

Examples:

    my_list = [1, 2, 3]

    print(2 in my_list)      # True
    print(4 not in my_list)  # True

7. Identity Operators
Check if two variables refer to the same object in memory.

Examples:

    a = [1, 2]
    b = a
    c = [1, 2]

    print(a is b)      # True (same object)
    print(a is c)      # False (same values, different objects)
    print(a is not c)  # True

#5- Explain the concept of type casting in Python with examples.
 - Type casting (also called type conversion) is the process of converting one data type into another. This is useful when you want to perform operations that require a certain type — for example, converting a string to an integer before doing math.


There are two types of casting in Python:

Implicit Type Casting – done automatically by Python

Explicit Type Casting – done manually using functions like int(), float(), str(), etc

1. Implicit Type Casting :
Python automatically converts smaller or compatible data types into larger ones during expressions.

Example:  x = 5   #int    
         int y = 2.5     # float

result = x + y

print(result)       # Output: 7.5

print(type(result))

 2. Explicit Type Casting:
You manually convert data types using functions like:

int() – converts to integer

float() – converts to float

str() – converts to string

bool() – converts to Boolean

list(), tuple(), set() – convert to collections

Example:

  num_int = int(num_str)

 1. for Loop :
    The for loop is used to iterate over a sequence (like a list, tuple, string, or range). It's ideal when you know in advance how many times you want to repeat something.

 Example:

   fruits = ["apple", "banana", "cherry"]

   for i in fruits:

   print(i)

Example:

count = 0

   Count is: 0
